We will be creating our own test data for analysis.

```python
# set seed
np.seed(111)

# Function to generate test data
def CreateDataSet(Number=1):
    
    Output = []
    
    for i in range(Number):
        
        # Create a weekly (mondays) date range
        rng = pd.date_range(start='1/1/2012', end='12/31/2012', freq='W-MON')

        # Create random data
        data = np.randint(low=25,high=1000,size=len(rng))
        
        # Status pool
        status = [1,2,3]
        
        # Make a random list of statuses
        random_status = []
        for i in range(len(rng)):
            random_status.append(np.randint(low=0,high=len(status) ))
        
        # State pool
        states = ['GA','FL','fl','NY','NJ','TX']
        
        # Make a random list of states 
        random_states = []
        for i in range( len(rng) ):
            random_states.append(states[np.randint(low=0,high=len(states)) ] ) 

        Output.extend(zip(random_states, random_status, data, rng))
        
    return Output
```

Now that we have a function to generate our test data, lets create some data and stick it into a dataframe.

```python
dataset = CreateDataSet(4)
df = pd.DataFrame(data=dataset, columns=['State','Status','CustomerCount', 'StatusDate'])
df.info()
```

```python
#Strip B off of a column with B's in them. 
companies = ["A","B","C","D"]
assets = ["3.2B","4.5B","1.2B","10.9B"]

dataset = list(zip(companies, assets))
billion = pd.DataFrame(data=dataset, columns=['Company','Assets'])
billion
```

```python
# Add a new column that removes the last letter of the first column.
billion["Assets_Number"] = billion['Assets'].apply(lambda x: x[:-1])
billion = billion.sort("Assets_Number")
billion

```

```python
# You actually want Assests_Number to be a float before sorting. 
billion["Assets_Number"] = billion['Assets_Number'].apply(lambda x: float(x))
billion = billion.sort("Assets_Number")
billion
```
